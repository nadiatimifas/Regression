```python
# importer les laibraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
```

```python
# importer le dataset
dataset = pd.read_excel('Data_Temp.xlsx')
```

```python
X=dataset.iloc[:,:-1]
Y=dataset.iloc[:,-1]
```

```python
dataset= dataset[dataset["Year"]<=2200]
```

```python
plt.scatter(X,Y,color="yellow")
```

```python
# crée et entrainé un modèle de regression linéaire simple
regressor = LinearRegression()
regressor.fit(X,Y)
Y_pred = regressor.predict(X)
```

```python
plt.scatter(X,Y,color='yellow')
plt.plot(X,regressor.predict(X),color="red")
```

```python
# Calculer la performance du modele 
r2 = r2_score(Y,Y_pred)
r2
```

```python
# Créer et entraine un modele de la regression polynomiale  
poly_reg = PolynomialFeatures(degree = 10 )
X_poly = poly_reg.fit_transform(X)
regressor_poly = LinearRegression()
regressor_poly.fit(X_poly,Y)
Y_poly_pred = regressor_poly.predict(X_poly)
```

```python
plt.scatter(X,Y,color="yellow")
plt.plot(X, regressor.predict(X), color = "red")
plt.plot(X, regressor_poly.predict(X_poly), color = "green")
```

```python
r2_poly = r2_score(Y,Y_poly_pred)
r2_poly
```

```python
# une nouvelle prédiction
predictions = poly_reg.fit_transform([[2030]])
regressor_poly.predict(predictions)
```
